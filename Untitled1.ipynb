```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
corn = Path("Resources/corn.csv")
corn_df = pd.read_csv(corn, index_col="Date Time",parse_dates=True, infer_datetime_format=True, skiprows=1)


corn_df = corn_df.sort_index()
#select column
corn_df = corn_df[["Close"]]
#rename column
corn_df = corn_df.rename(columns={'Close': 'corn'})
```

```python
# Count nulls
corn_df.isnull().sum().sum()
# Drop nulls
corn_df = corn_df.dropna().copy()
```

```python
corn_df = corn_df.iloc[:-1,:]
```

```python
corn_df
```
