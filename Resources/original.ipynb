```python
# Read in the csv and pre-process all date time column to be standadrd (replace / by -)
```

```python
file_list = ['ada.csv','amc.csv','bb.csv','btc.csv','corn.csv','crown.csv','doge.csv','dominos.csv','eth.csv','gme.csv','hvn.csv','oil.csv','pmv.csv','tsla.csv','vix.csv','wheat.csv']

df_indicators = pd.DataFrame()
for file in file_list:
    
    def add_padding_hour(date_time):
        print(date_time)
        split_date_time = date_time.split(' ')
        print(split_date_time)
        if int(split_date_time[1][0]) < 11:
            return split_date_time[0] + ' 0' + split_date_time[1]
        else:
            return date_time
        
    #df_ind = pd.read_csv(file, index_col="Date Time",parse_dates=True, infer_datetime_format=True, skiprows=1)
    df_ind = pd.read_csv(file, skiprows=1).iloc[:-1,:]
    if "/" in df_ind["Date Time"][0]:

        #df_ind.index = pd.to_datetime(df_ind.index, format = '%d/%m/%Y %H:%M')
        #df_ind.index = df_ind.index.to_series().apply(add_padding_hour)

        df_ind["Date Time"] = pd.to_datetime(df_ind["Date Time"], format = '%d/%m/%Y %H:%M')
    else:
        df_ind["Date Time"] = pd.to_datetime(df_ind["Date Time"],format = '%Y-%m-%d %H:%M:%S')
        
        
    # convrt Date Time column into datetime type
    
    df_ind['product'] = file.split('.csv')[0]
    df_ind['pct change'] = df_ind[["Close"]].pct_change()
   # df_ind['stdev'] = df_ind[["Close"]].std()
    
    df_ind = df_ind[["Date Time","product",'Close','pct change']]
    
    df_ind = df_ind.iloc[:-1,:]
    df_ind = df_ind.dropna().copy()
    
    #df_ind.index = pd.to_datetime(df_ind.index, format = '%d/%m/%Y %H:%M')
    #df_ind.index =df_ind.index.strftime('%Y-%m-%d %H:%M')
    
    #clean it up
   
    df_indicators = pd.concat([df_indicators,df_ind],axis='rows')
df_indicators.info
```

```python
df_indicators.hvplot.line(
    "Date Time", 
    "Close",
    xlabel="Date Time", 
    ylabel="Close",
    rot=90,
    height = 200,
    color='black',
    colormap='pink',
    groupby='product')
```
